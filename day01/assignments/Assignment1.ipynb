## Exercice 1

```python
import pandas as pd

dataO0 = pd.read_csv('Mean/resultsO0.txt',names=['stride','mean','millis','rate'])
dataO2 = pd.read_csv('Mean/resultsO2.txt',names=['stride','mean','millis','rate'])
dataO0.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(dataO0['stride'],dataO0['millis'],'o-',c='r',label='using -O0')
plt.plot(dataO2['stride'],dataO2['millis'],'o-',c='b',label='using -O2')
plt.xlabel('Stride')
plt.ylabel('CPU time, ms')
plt.legend()
```

```python
plt.figure(figsize=(12,8))
plt.plot(dataO0['stride'],dataO0['rate'],'o-',c='r',label='using -O0')
plt.plot(dataO2['stride'],dataO2['rate'],'o-',c='b',label='using -O2')

plt.xlabel('Stride')
plt.ylabel('Effective Bandwith, MB/s')
plt.legend()
```

#### Conclusion :
Au fur et à mesure que le nombre de stride augmente le temps cpu augmente tandis que la bande passante diminue. Cela est normal car avec les strides les données à manipulées ne sont plus contigues dans le tableau et il faut faire des "sauts" pour aller chercher les données à chaque fois. Du coup plus la stride est grande, plus le programme met du temps d'une case à l'autre. En outre cela entraîne une diminution du débit des opérations de lecture-écriture d'où la diminution de la bande passante

```python
datablock = pd.read_csv('mxm_matrice/results_matrix_block.txt',names=['Block','millis','rate'])
datablock.head()
```

```python
plt.figure(figsize=(12,8))
plt.plot(datablock['Block'],datablock['millis'],'o-',c='b',label='using -O2')
plt.xlabel('Block')
plt.ylabel('CPU time, ms')
plt.legend()
plt.title('MxM Block Version')
```

#### Conclusion : 
La meilleure taille de block est 50 car c'est pour cette taille que nous avons le CPU times le plus bas et la bande passante la plus grande.
